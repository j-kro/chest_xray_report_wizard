```python
#@title

import ipywidgets as widgets
from IPython.display import display

# Vergleichsuntersuchung

##Definition der Variablen
kvgl = 'Keine Vergleichsuntersuchung vorliegend.'

rt = 'Zum Vergleich liegt zuletzt ein Thoraxröntgen vom '
ct = 'Zum Vergleich liegt zuletzt eine CT des Thorax vom '
rct = 'Zum Vergleich liegen zuletzt ein Thoraxröntgen und eine CT des Thorax vom '
pet = 'Zum Vergleich liegt zuletzt eine PET/CT vom '

vergleich = widgets.HTML(value="<i>Vergleich:</i>")

##Vegleich RadioButtons
vgl = widgets.RadioButtons(
    options=['Keine Vergleichsuntersuchung', 'Vergleichsuntersuchung vom Vortag', 'Vergleichsuntersuchung vom:'],
    value='Vergleichsuntersuchung vom:',
    disabled=False
)

##Datumsfeld
date = widgets.Text(
    layout={'width': '180px'},
    placeholder='Datum',
    #description='',
    disabled=False
)

##Vergleichsmodalität
vmod = widgets.HTML(value="<i>Vergleichsmodalität:</i>")

mod = widgets.RadioButtons(
    options=[('Röntgen', rt), ('CT', ct), ('Röntgen und CT', rct), ('PET/CT', pet)],
    layout={'width': 'max-content'},
    disabled=False
)

##Vergleich als HBox
vgl_items = [vergleich, vgl, date, vmod, mod]
left_box1 = widgets.VBox([vgl_items[0], vgl_items[1], vgl_items[2]])
right_box1 = widgets.VBox([vgl_items[3], vgl_items[4]])
vgl_items_box = widgets.HBox([left_box1, right_box1])

#'Befund und Beurteilung' einfügen
#bub = 'Befund und Beurteilung'

#bub_check = widgets.Checkbox(
#    layout={'width': 'max-content'},
#    description= bub,
#    value=False
#)


# Lagewechsel
st = 'Lagewechsel zum Stehen.'
lg = 'Lagewechsel zum Liegen.'

lage = widgets.RadioButtons(
    options=[('kein Lagewechsel',''), ('zum Stehen', st), ('zum Liegen', lg)],
    description='Lagewechsel:',
    disabled=False
)

# Nur eine Ebene als check box
ebene = 'Aufnahme in einer Ebene.'

ebene_check = widgets.Checkbox(
    layout={'width': 'max-content'},
    description=ebene,
    value=False
)

##HBox von Lage und 1 Ebene
lg_items = [lage, ebene_check]
lg_items_box = widgets.HBox(lg_items)

# Befundkörper

## Standardtext als radiobuttons
nb = 'Kein Pneumothorax. Keine Pleuraergüsse. Keine infiltrattypischen Verschattungen. Keine höhergradige pulmonalvenöse Stauung. Herzschatten nicht verbreitert. Oberes Mediastinum schlank. Retrosternal- und Retrokardialraum frei einsehbar. Erhaltenes Wirbelsäulenalignment.'
kbw = 'Projektionsradiographisch kein durchgreifender Befundwandel.'
ft = ''

bef = widgets.RadioButtons(
    options=[('Kein Befundwandel', kbw), ('Normalbefund', nb), ('Nur Freitext', ft)],
    placeholder='Freitext',
    description='Befund:',
    disabled=False
)

text = widgets.Textarea(
    layout={'height':'auto', 'width':'650px'},
    value='',
    placeholder='Freitext',
    description='Freitext:',
    disabled=False
)

## Optionale Befundelemente

### Im Übrigen
imue = 'Im Übrigen kein durchgreifender kardiopulmonaler Befundwandel.'

imue_check = widgets.Checkbox(
    layout={'width': 'max-content'},
    description='Im Übrigen kein Befundwandel',
    value=False
)

### Kein Pneu etc.
kpneu = 'Kein Pneumothorax. Keine Pleuraergüsse. Keine Infiltrate.'

kpneu_check = widgets.Checkbox(
    layout={'width': '330px'},
    description='Kein Pneu. Keine Ergüsse. Kein Infitrat.',
    value=False
)
### Mediastinum
hs = 'Herzschatten nicht verbreitert.'

hs_check = widgets.Checkbox(
    #layout={'width': 'max-content'},
    description=hs,
    value=False
)

### Wirbelsäule
ws = 'Erhaltenes Wirbelsäulenalignment.'

ws_check = widgets.Checkbox(
    layout={'width': 'max-content'},
    description=ws,
    value=False
)

### Fremdmaterial
fremd = 'Fremdmaterial, idem.'

fremd_check = widgets.Checkbox(
    #layout={'width': 'max-content'},
    description=fremd,
    value=False
)

iu_fremd = 'Sonstiges Fremdmaterial, idem.'

iu_fremd_check = widgets.Checkbox(
    layout={'width': 'max-content'},
    description=iu_fremd,
    value=False
)

###Layout optionale Befundelemente als Box

items = [imue_check, kpneu_check, hs_check, ws_check, fremd_check, iu_fremd_check]
left_box = widgets.VBox([items[0], items[1], items[2]])
right_box = widgets.VBox([items[3], items[4], items[5]])
elemente = widgets.HBox([left_box, right_box])

### Optionale Elemente Freitext
opt_text = widgets.Textarea(
    layout={'height':'auto', 'width':'650px'},
    value='',
    placeholder='Freitext',
    description='Freitext:',
    disabled=False
)

# CT Befund zum Vergleich
ct_vgl = 'Bezüglich des ausführlichen Befundes verweisen wir auf die CT des Thorax vom gleichen Tag.'

ct_vgl_check = widgets.Checkbox(
    layout={'width': 'max-content'},
    description='CT-Befund vom gleichen Tag vorliegend',
    value=False
)

# Befundmitteilung

kein = 'Keine Befundmitteilung'
spkoll = 'Befundmitteilung an folgenden Kollegen:'
mkoll = 'Befundmitteilung an den behandelnden Kollegen'
wkoll = 'Befundmitteilung an die behandelnde Kollegin'

drop = widgets.Dropdown(
    options=[kein, mkoll, wkoll, spkoll],
    value=kein,
    description='Mitteilung:',
    disabled=False,
)

name = widgets.Text(
    #value=,
    placeholder='Name',
    description='Kliniker:',
    disabled=False
)

# Eingabemaske erstellen.

##Vergleich
display(widgets.HTML(value="<b>Chest x-ray report wizard</b>"))
display(vgl_items_box)
print('')

##Befundkörper
display(widgets.HTML(value="<b>Befund</b>"))
display(lg_items_box)

display(widgets.HTML(value="<i>Standardbefund auswählen oder Freitext-Befund schreiben:</i>"))
display(bef)
display(text) 
print('')

##Optionale Befundelemente
display(widgets.HTML(value="<i>Optionale Befundelemente:</i>"))
display(elemente)
display(opt_text)
print('')

##CT zum Vergleich
display(widgets.HTML(value="<i>CT zum Vergleich:</i>"))
display(ct_vgl_check)
print('')

##Befundmitteilung
display(widgets.HTML(value="<b>Befundmitteilung</b>"))
display(drop)
display(name)
print('')

# Befund erstellen.
def tb():
  if type(opt_text.value) == str and type(opt_text.value) != '':
    print(opt_text.value)


def befundtext():
  if vgl.value == 'Keine Vergleichsuntersuchung':
    print(kvgl, lage.value) 
  elif vgl.value == 'Vergleichsuntersuchung vom Vortag':
    print(mod.value + 'Vortag vor.', lage.value)
  else:
    print(mod.value+date.value+' vor.', lage.value)
  if ebene_check.value == True:
      print(ebene)
  if bef.value == ft:
    print(text.value)
  else:
    print(bef.value)
    print(text.value)
  if imue_check.value == True: 
    print(imue)
  if kpneu_check.value == True: 
    print(kpneu)
  if hs_check.value == True:
    print(hs)
  if ws_check.value == True: 
    print(ws)
  if fremd_check.value == True: 
    print(fremd)  
  if iu_fremd_check.value == True:
    print(iu_fremd)
  if type(opt_text.value) == str and type(opt_text.value) != '':
    print(opt_text.value)
  if ct_vgl_check.value == True: 
    print(ct_vgl)  
  if drop.value == mkoll:
    print(drop.value+' erfolgt.')
  if drop.value == wkoll: 
    print(drop.value+' '+name.value+'erfolgt.')
  if drop.value == spkoll:
    print('Befundmitteilung an '+name.value+' erfolgt.') 
  print('')


## Button für die Befunderstellung
button = widgets.Button(description="Befund erstellen")
button.style.button_color='lightblue'
output = widgets.Output()

display(button, output)

def on_button_clicked(b):
    with output:
        output.clear_output()
        print('')
        print('-------------------------------------------------------------------')
        befundtext()
button.on_click(on_button_clicked)
```
